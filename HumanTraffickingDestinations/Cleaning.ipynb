```python
import pandas as pd
import numpy as np
import unicodecsv
import seaborn as sns
import matplotlib.pyplot as plt
import pprint 
%matplotlib inline

data = pd.read_csv('HTI.csv')
```

```python
data.keys()
```

```python
# Determine if there are any null values
data.isnull().sum()
```

### Filter to just country, year and type

```python
# Create a filtered data set that includes country, year, and the seven labor types
df = pd.DataFrame(data)
df = data[['country', 'year', "pdest", "ldest", "ddest", "dsdest", "cpdest", "cldest", "csdest"]]
df[0:5]
```

### Boolean values for labor types

```python
# Boolean transform 
d = {'Yes': True, 'No mention': False, 'No': False}
df = df.replace(d)
df[0:5]
```

### Fill missing years


```python
# Define function to fill in null data points for missing years
def country_years_missing(dataframe):
    years = list(range(2001,2012))
    country = list(df.country.unique())
    for x in country:
        for y in years: 
            if dataframe.query("year == @y and country == @x").empty:
                d = {'country': [x], 'year': [y]}
                dataframe = dataframe.append(pd.DataFrame(d), ignore_index=True) 
    
    return dataframe
```

```python
# Fill in null data points for missing years
df = country_years_missing(df)
```
